```python
class Circle():
    pi = 3.14

    def __init__(self,rad):
        self.radius = rad
    
    def area(self):
        return self.radius* self.radius * Circle.pi
    def setRadius(self, rad):
        self.radius = rad
    def getRadius(self):
        return self.radius
        
```

```python
c = Circle(rad = 4)
```

```python
c.area()
```

```python
c.setRadius(2)
```

```python
c.area()
```

```python
c.setRadius(5)
```

```python
c.getRadius()
```

class level attributes, Instance level attributes, static attributes

```python
class School():
    SchoolName = "ABC School"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def student_details(self):
        return self.name, self.age, School.SchoolName
    @classmethod
    def changeSchoolName(cls, name):
        cls.SchoolName = name
    @staticmethod
    def notes(Subject):
        return ["Chapter1", "Chapter2"]
```
