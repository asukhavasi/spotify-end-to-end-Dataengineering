Pivot table

```python
d = {
    'A':['food','food','food','bar','bar','bar'],
    'B':['x','y','x','y','x','y'],
    'C':[1,2,3,4,5,1]
}
df = pd.DataFrame(d)
```

```python
d_pivot = df.pivot_table(
    values= 'C',
    index='A',
    columns=['B'],
    aggfunc = np.sum
)
```

```python
d_pivot
```

```python
d_pivot.reset_index()
```

```python
df
```

```python
df1 = df.groupby(['A','B']).agg(np.sum).reset_index()
```

```python
df1
```

Concat, Merge, Join

```python
df1 = {
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
}
```

```python
df1
```

```python
df1 = pd.DataFrame(df1,index=['K0','K1','k2'])
```

```python
df1
```

```python
df2 = pd.DataFrame({
    'A':['A0','B0','C0'],
    'C':['A0','A1','A2'],
    'D':['B0','B1','B2'],
},index=['K0','K3','K1'])
```

```python
df2
```

```python
df = df1.join(df2, how='inner')
```

```python
df
```

```python
pd.merge(df1, df2, how='left', on=['A'])
```

```python
pd.merge(df1,df2,how='right',on=['A'])
```
